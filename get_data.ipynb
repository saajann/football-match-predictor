```python
import requests
import pandas as pd
```

```python
API_URL = 'https://api.football-data.org/v4/competitions/SA/teams?season=2024'
API_KEY = ''

headers = {'X-Auth-Token': API_KEY}
response = requests.get(API_URL, headers=headers)
```

```python
data = response.json()

teams = pd.json_normalize(data['teams'])
teams = teams[['id', 'name']]

teams.to_csv('team_ids.csv', index=False)
```

```python
import time
matches_list = []
i = 0;

for team_id in teams['id']:
    API_URL = f'https://api.football-data.org/v4/teams/{team_id}/matches?season=2024'
    response = requests.get(API_URL, headers=headers)
    matches = response.json()['matches']
    matches_list.extend(matches)
    i += 1
    if i == 9:
        time.sleep(60)
```

```python

matches_df = pd.json_normalize(matches_list)

matches_df.to_csv('matches.csv', index=False)
```
