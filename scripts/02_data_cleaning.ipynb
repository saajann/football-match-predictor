```python
import pandas as pd
```

```python
df = pd.read_csv('../data/serie_a_matches.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['area', 'competition', 'season', 'stage', 'lastUpdated', 'odds', 'group', 'referees'], axis=1, inplace=True)
```

```python
df['date_month'] = df['utcDate'].str.split('T').str[0].str[5:7]
df['time_hour'] = df['utcDate'].str.split('T').str[1].str.replace('Z', '').str[:2]
```

```python
df.drop('utcDate', axis=1, inplace=True)
```

```python
df.head()
```

```python
import ast

def extract_short_name(team_str):
    team_dict = ast.literal_eval(team_str)
    return team_dict['shortName']

df['homeTeam'] = df['homeTeam'].apply(extract_short_name)
df['awayTeam'] = df['awayTeam'].apply(extract_short_name)
```

```python
df.head()
```

```python
def fulltime_home(team_str):
    team_dict = ast.literal_eval(team_str)
    return team_dict['fullTime']['home']
def fulltime_away(team_str):
    team_dict = ast.literal_eval(team_str)
    return team_dict['fullTime']['away']
def halftime_home(team_str):
    team_dict = ast.literal_eval(team_str)
    return team_dict['halfTime']['home']
def halftime_away(team_str):
    team_dict = ast.literal_eval(team_str)
    return team_dict['halfTime']['away']
```

```python
df['ft_home'] = df['score'].apply(fulltime_home)
df['ft_away'] = df['score'].apply(fulltime_away)
df['ht_home'] = df['score'].apply(halftime_home)
df['ht_away'] = df['score'].apply(halftime_away)
```

```python
df.loc[0]['score']
```

```python
df.drop('score', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['winner'] = df.apply(lambda row: 0 if row['ft_home'] > row['ft_away'] else (1 if row['ft_home'] == row['ft_away'] else 2), axis=1)
```

```python
df.to_csv('../data/serie_a_matches_cleaned.csv', index=False)
```
